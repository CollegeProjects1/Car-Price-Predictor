```python
dataset1.info()
backup=dataset1.copy()
data=pd.DataFrame(dataset1)
```

```python
data=data.drop(['Source.Name','web-scraper-order','web-scraper-start-url'],axis=1)
data.head()
data.info()
```

```python
from numpy import NaN
data.dropna(inplace=True)
data.info()
data.head()
```

```python
data['selling_price']=data['selling_price'].str.split(' ').str.get(0)
```

```python
data.head()
```

```python
data['km_driven']=data['km_driven'].str.split(' ').str.get(0)
data['km_driven']=data['km_driven'].str.replace(',','')
data.head()
```

```python
mileage=data['mileage'].str.split(' ').str.get(0)
data['mileage']=mileage.str.slice(7,)
data.head()
```

```python
max_power=data['max_power'].str.split(' ').str.get(1)
data['max_power']=max_power.str.slice(5,)
data.head()
```

```python
engine=data['engine'].str.split(' ').str.get(0)
data['engine']=engine.str.slice(6,)
data.head()
```

```python
newprice=data['new-price'].str.split('-').str.get(2)
newprice=newprice.str.split(' ').str.get(0)
data['new-price']=newprice
data.head()
```

```python
data.info()
```

```python
data=data.drop(['seats'],axis=1)
data=data.drop(['engine','max_power'],axis=1)
```

```python
data.head()
```

```python
data['selling_price']=data['selling_price'].str.replace('*','')
data['selling_price']=data['selling_price'].str.replace(',','.')
data.info()
```

```python
data['selling_price']=data['selling_price'].astype(float)
data['new-price']=data['new-price'].astype(float)
data['km_driven']=data['km_driven'].astype(float)
data['mileage']=data['mileage'].astype(float)
data.info()
```

```python
data['full_name']=data['full_name'].str.split(' ').str.slice(0,3).str.join(' ')
data['Brand']=data['full_name'].str.split(' ').str.get(0)
print()
data.info()
data.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
mpl.style.use('ggplot')
import seaborn as sns
```

```python
plt.subplots(figsize=(15,7))
ax=sns.boxplot(x='Brand',y='selling_price',data=data)
ax.set_xticklabels(ax.get_xticklabels(),rotation=40,ha='right')
plt.show()
```

```python
plt.subplots(figsize=(15,7))
ax=sns.swarmplot(x='year',y='selling_price',data=data)
ax.set_xticklabels(ax.get_xticklabels(),rotation=40,ha='right')
plt.show()
```
